## 1. Sustainable Lifestyle Rating Dataset

```python
sustainableLifestyle_df = sustainableLifestyle_df.drop(
    columns=[
        col for col in sustainableLifestyle_df.columns
        if col not in ['DietType', 'LocalFoodFrequency', 'TransportationMode', 'Rating']
        ]
    )

print(sustainableLifestyle_df)
```

## 2. Individual Carbon Emison Dataset

```python
idvCarbonEmision_df.info()
```

```python
idvCarbonEmision_df.duplicated().sum()
```

```python
idvCarbonEmision_df = idvCarbonEmision_df.drop(
    columns=[
        col for col in idvCarbonEmision_df.columns
        if col not in ['Diet', 'Transport', 'Vehicle Type ', 'Frequency of Traveling by Air', 'Vehicle Monthly Distance Km', 'CarbonEmission']
        ]
    )

print(idvCarbonEmision_df)
```

## 3. Uni Student Responses Dataset

```python
uniStudentResponses_df = pd.read_csv('data/UniStudentResponses.csv')
uniStudentResponses_df.head()
```

```python
uniStudentResponses_df.info()
```

```python
uniStudentResponses_df.duplicated().sum()
```

```python
uniStudentResponses_df.describe()
```

```python
uniStudentResponses_df = uniStudentResponses_df.drop(
    columns=[
        col for col in uniStudentResponses_df.columns
        if col not in ['How do you travel to work/university?', 'On average, how long does it take you to travel to work/university?']
        ]
    )

uniStudentResponses_df = uniStudentResponses_df.rename(columns={
    'How do you travel to work/university?': 'transportationToTravel',
    'On average, how long does it take you to travel to work/university?': 'avarageDistanceToTravel'
})

print(uniStudentResponses_df)
```
